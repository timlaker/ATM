```python
from IPython.display import clear_output
clear_output()

# Function to create a board to display to user
def display_board(board):
    clear_output() # Clears the previous board everytime (only works in jupyter files)
    print(board[1]+' | '+board[2]+' | '+board[3])
    print('---------')
    print(board[4]+' | '+board[5]+' | '+board[6])
    print('---------')
    print(board[7]+' | '+board[8]+' | '+board[9])

board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']


# Function for player to choose their marker
def player_input():

    marker = ''
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1: Do you want to be X or O? ').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')


# Function to enable the player to place a marker to a position on the board
def place_marker(board, marker, position):
    # The board @ chosen position = marker placement
    board[position] = marker


# Function to check all possible winning outcomes
def win_check(board,mark):

    return ((board[7] == mark and board[8] == mark and board[9] == mark) or 
    (board[4] == mark and board[5] == mark and board[6] == mark) or 
    (board[1] == mark and board[2] == mark and board[3] == mark) or 
    (board[7] == mark and board[4] == mark and board[1] == mark) or 
    (board[8] == mark and board[5] == mark and board[2] == mark) or 
    (board[9] == mark and board[6] == mark and board[3] == mark) or 
    (board[7] == mark and board[5] == mark and board[3] == mark) or 
    (board[9] == mark and board[5] == mark and board[1] == mark)) 

import random

# Function to randomly choose which player goes first, if random number return is 0, player 2 is first..
def choose_first():
    if random.randint(0, 1) == 0:
        return 'Player 2'
    else:
        return 'Player 1'


# Function to check wether there are any empty spaces (strings) on the board still
def space_check(board,position):

    return board[position] == ' '


# Function to check if the board is full, returns False if not full, True if so.
def full_board_check(board):

    for i in range(1,10):
        if space_check(board,i): # if space_check returns True (spaces left) then return False here.
            return False
    return True

# Function to allow player to choose where to place a marker
def player_choice(board):

    position = 0 # Set at zero as nothing at position 0 for this game
    # While out of range available OR space not available, keep asking for position from the user.
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        position = int(input('Choose a position: (1-9) '))
    # Otherwise, return position
    return position

# Function to allow user to play again or not.
def replay():
    
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')


# Setting up the game itself using the above functions.
print('Welcome to Tic Tac Toe!')


while True:
    theBoard = [' '] * 10 # Reset the board - 10 blank strings
    player1_marker, player2_marker = player_input() # Define player markers by player input function
    turn = choose_first() # define turn as choose_first function to randomise
    print(turn + ' will go first.') # Print who goes first for user
    
    play_game = input('Are you ready to play? Enter Yes or No.') # Ask for user input to play game or not
    
    if play_game.lower()[0] == 'y':
        game_on = True # if first letter the user types starts with y, game_on = True
    else:
        game_on = False

    while game_on: # While game_on as above is True..
        if turn == 'Player 1':
            
            # Player1's turn.
            
            display_board(theBoard) # Pass theBoard as defined above into display_board function
            position = player_choice(theBoard) # Player chooses their position on theBoard
            place_marker(theBoard, player1_marker, position) # Passes board, marker1 and chosen position into place_marker function.

            if win_check(theBoard, player1_marker): # Runs win_check function after each placement
                display_board(theBoard)
                print('Congratulations! You have won the game!')
                game_on = False # Ends game and prints the above if someone won
            else:
                if full_board_check(theBoard): # If full_board = True, tied game message printed
                    display_board(theBoard)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 2' # otherwise, continue to next player

        else:
            # Player2's turn - same checks as above then pass to player 1
            
            display_board(theBoard)
            position = player_choice(theBoard)
            place_marker(theBoard, player2_marker, position)

            if win_check(theBoard, player2_marker):
                display_board(theBoard)
                print('Player 2 has won!')
                game_on = False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('The game is a draw!')
                    break
                else:
                    turn = 'Player 1'

    if not replay(): # If user chooses not to continue playing, break the while loop.
        break

         
```
